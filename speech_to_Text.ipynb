```python
# Install required packages
!pip install librosa soundfile noisereduce transformers torch torchaudio pydub --quiet

# Step 1: Use the uploaded file
input_file = "WhatsApp Audio 2025-10-01 at 12.37.28_db464187.dat.unknown"
output_wav = "input.wav"

from pydub import AudioSegment

# Convert any audio format to WAV
audio = AudioSegment.from_file(input_file)
audio.export(output_wav, format="wav")
print(f"Converted to WAV: {output_wav}")

# Step 2: Clean the audio
import librosa, soundfile as sf, noisereduce as nr

def clean_audio(input_file, output_file="final_cleaned.wav"):
    y, sr = librosa.load(input_file, sr=16000, mono=True)
    y = librosa.util.normalize(y)
    y = nr.reduce_noise(y=y, sr=sr)
    y, _ = librosa.effects.trim(y, top_db=20)
    sf.write(output_file, y, sr)
    print(f"Cleaned audio saved as {output_file}")

clean_audio(output_wav, "final_cleaned.wav")

# Step 3: Speech-to-text using Whisper (long audio safe)
from transformers import pipeline

asr = pipeline(
    "automatic-speech-recognition",
    model="openai/whisper-small",
    chunk_length_s=30,       # split long audio automatically
    return_timestamps=True    # required for audio >30s
)

def speech_to_text(audio_file):
    result = asr(audio_file)
    full_text = " ".join([seg["text"] for seg in result["chunks"]])
    return full_text

text = speech_to_text("final_cleaned.wav")
print("Recognized Text:", text)
```
